```python
# Creer une fonction
def nb_entiers_hazard(debut=0, fin=100, longeur=1000):
    """
    Voici ce que la fonction fait:
    
    - Fait une liste, de longeur egal au 
      parametre longeur, de nombres choisi 
      au hazard entre les parametres debut 
      et fin.
    
    Parametres
    -------------------------------
    debut: entier
    
        C'est le chiffre plus bas possible
        
    fin: entier
    
        Plus haut possible.
        
    longeur: entier
    
        Longeur de liste ordonnee.
        
    Retour
    --------------------------------
    output: npliste
    
        Redonne la liste cree.
    """
    print('Fonction: random_array(debut={}, fin={}, longeur={})'.format(debut, fin, longeur))
    return np.random.randint(debut, fin+1, longeur)
```

```python
# ------------------------ Tester nb_entiers_hazard ------------------------------

# Avec defaults:
default_entiers = nb_entiers_hazard()
print('Avec defaults: \n', default_entiers, '\n\n')
      
# Avec parametres choisis:
hazard_entiers = nb_entiers_hazard(100, 101, 5)
print('Avec parametres choisis: \n', hazard_entiers, '\n\n')
```

```python
def combien_de_pattates_pillees(liste, save=False):
    
    # Creer un liste vide de meme taille
    pattates = [' ' for i in range(len(liste))]
    
    # Lire les chiffres de la fin de la liste
    # jusqu'au debut.
    for indexe in range(len(liste)):
        
        if liste[indexe]==0:
            # Si c'est 0, dis aucune pattate pillee
            pattates[indexe] = 'Aucune pattate pillee!'
            print(pattates[indexe])
            
        elif liste[indexe]==1:
            # Si c'est 1, dis 1 pattate pillee
            pattates[indexe] = str('{} patatte pillee!'.format(liste[indexe]))
            print(pattates[indexe])
            
        elif liste[indexe]==2:
            # Si c'est 2, dis une paire de pattates pillees
            pattates[indexe] = str('{} paire de patattes pillees!'.format(int(liste[indexe]/2)))
            print(pattates[indexe])

        elif liste[indexe]%2==0:
            # Sinon si c'est divisible par 2 et pas 0, dis paires de pattates
            pattates[indexe] = str('{} paires de patattes pillees!'.format(int(liste[indexe]/2)))
            print(pattates[indexe])
        
        else:
            # Sinon si pas divisible par 2 mais pas 0, dis juste pattates
            pattates[indexe] = '{} patattes pillees!'.format(liste[indexe])
            print(pattates[indexe])
    
    if save==True:
        pattates = np.array([item for item in pattates if not item==' '])
        print('Saving...')
        np.savetxt('liste_de_pattates.txt', pattates, fmt='%s', delimiter='\n')
        print('Saved')
```

```python
# ------------------------ Tester combien_de_pattates_pillees ------------------------------

# Avec defaults:
print('Avec defaults: \n')
combien_de_pattates_pillees(default_entiers, save=True)
```
